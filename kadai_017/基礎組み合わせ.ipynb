```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(self.name, "は大人です")
        else:
            print(self.name, "は大人ではありません")


# インスタンスを複数作成
human1 = Human("太郎", 25)
human2 = Human("花子", 18)
human3 = Human("次郎", 30)

# リストに追加
humans = [human1, human2, human3]

# リストの要素数分だけメソッドを呼び出す
for person in humans:
    person.check_adult()
```
