```python
# 上辺・下辺・高さを変数に代入
upper = 10
lower = 20
height = 5

# 台形の面積を計算して変数に代入
area = (upper + lower) * height / 2

# 結果を単位付きで出力
print(str(area) + "cm²")
```

```python
# 上辺・下辺・高さを変数に代入
upper = 10
lower = 20
height = 5

# 台形の面積を計算して変数に代入
area = (upper + lower) * height / 2

# 結果を単位付きで出力
print(f"{area}cm²")
```
